Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Scatter Plot:

Use: Visualizes the relationship between two continuous variables. It is useful for spotting correlations or clusters in data.
Example Function: sns.scatterplot()


Line Plot:

Use: Displays trends over a continuous interval or time series data. It is commonly used to observe changes or trends over time.
Example Function: sns.lineplot()


Box Plot:

Use: Summarizes the distribution of a dataset and identifies outliers. It displays the median, quartiles, and potential outliers.
Example Function: sns.boxplot()


Histogram:

Use: Shows the frequency distribution of a single continuous variable, helping to understand the dataâ€™s distribution, shape, and central tendency.
Example Function: sns.histplot()


Heatmap:

Use: Represents data in a matrix format using colors. It is ideal for visualizing correlation matrices or pivoted datasets to detect patterns.
Example Function: sns.heatmap()

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri = sns.load_dataset('fmri')

# Option 1: Using hue and style in a single plot:
sns.lineplot(data=fmri, x='timepoint', y='signal', hue='event', style='region')
plt.title("FMRI Signal over Time by Event and Region")
plt.show()

# Option 2: Using facetting by region:
g = sns.FacetGrid(fmri, col='region', hue='event', height=4, aspect=1.2)
g.map(sns.lineplot, 'timepoint', 'signal')
g.add_legend()
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the titanic dataset
titanic = sns.load_dataset('titanic')

# Create a figure with two subplots
plt.figure(figsize=(12, 5))

# Box plot for age vs. pclass
plt.subplot(1, 2, 1)
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title("Age Distribution by Passenger Class")

# Box plot for fare vs. pclass
plt.subplot(1, 2, 2)
sns.boxplot(data=titanic, x='pclass', y='fare')
plt.title("Fare Distribution by Passenger Class")

plt.tight_layout()
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset
diamonds = sns.load_dataset('diamonds')

# Plot a histogram of the 'price' column using 'cut' as hue
sns.histplot(data=diamonds, x='price', hue='cut', multiple='stack')
plt.title("Distribution of Diamond Prices by Cut")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset('iris')

# Create a pair plot with 'species' as the hue
sns.pairplot(iris, hue='species')
plt.suptitle("Pair Plot of Iris Dataset", y=1.02)
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights = sns.load_dataset('flights')

# Pivot the dataset to get months as rows, years as columns, and passengers as values
flights_pivot = flights.pivot(index='month', columns='year', values='passengers')

# Plot the heatmap
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Number of Passengers Over Years and Months")
plt.show()
```
