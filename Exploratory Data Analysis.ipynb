Q1. Load the flight price dataset and examine its dimensions. How many rows and columns does the
dataset have?


```python
import pandas as pd

# Load the Excel file
file_path = "flight_price.xlsx"
df = pd.read_excel(file_path)

# Examine the dimensions
num_rows, num_cols = df.shape
print(f"The dataset has {num_rows} rows and {num_cols} columns.")
```

Q2.What is the distribution of flight prices in the dataset? Create a histogram to visualize the
distribution.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'df' with a 'flight_prices' column
# Load your data into the DataFrame or replace this with your data loading process

# Create a histogram
plt.figure(figsize=(10, 6))
plt.hist(df['Price'], bins=20, color='blue', edgecolor='black')

# Add labels and title
plt.xlabel('Flight Price')
plt.ylabel('Frequency')
plt.title('Distribution of Flight Prices')

# Show the plot
plt.show()
```
